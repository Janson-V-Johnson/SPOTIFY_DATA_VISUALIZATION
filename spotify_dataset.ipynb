```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
df = pd.read_csv('/content/dataset.csv')
```

```python
# CHECKING FOR MISSING VALUES #
df.isnull().sum()
```

```python
df["artists"].unique()
```

```python
# ANALYZING EACH DISTRIBUTIONS #

numeric_columns = ['popularity', 'duration_ms', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(4, 3, i)
    sns.histplot(df[column], bins=20, kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
# TOP 10 MOST FREQUENT ARTISTS #
popular_artists = df['artists'].value_counts().head(10)
print("Top 10 Most Frequent Artists:\n", popular_artists)
```

```python
# Top 10 Most Frequent Tracks #
popular_tracks = df['track_name'].value_counts().head(10)
print("Top 10 Most Frequent Tracks:\n", popular_tracks)
```

```python
# Top 10 LEAST Frequent Tracks #
popular_tracks = df['track_name'].value_counts().tail(10)
print("Top 10 less Frequent Tracks:\n", popular_tracks)
```

```python
# MOST POPULAR TRACK
popular_tracks = df.sort_values('popularity', ascending=False).head(1)

# Print the results
print(" Most Popular Tracks:")
for _, row in popular_tracks.iterrows():
    print(f"{row['track_name']}: {row['popularity']}")
```

```python
# LEAST POPULAR TRACK
least_popular_tracks = df.sort_values('popularity', ascending=True).head(1)

# Print the results
print("Least Popular Tracks:")
for _, row in least_popular_tracks.iterrows():
    print(f"{row['track_name']}: {row['popularity']}")
```

```python
#positive POPULARITY IN EACH GENRE #
genre_popularity = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10)
print("positive Popularity by Genre:\n", genre_popularity)
```

```python
#Negative  POPULARITY IN EACH GENRE #
genre_popularity = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).tail(10)
print("Negative Popularity by Genre:\n", genre_popularity)
```

```python
# AVERAGE DURATION
avg_duration = df['duration_ms'].mean()
print(f"Average Duration: {avg_duration:.2f} ms")
```

```python
#AVERAGE TEMPO
avg_tempo = df['tempo'].mean()
print(f"Average Tempo: {avg_tempo:.2f} BPM")
```

```python
#MOST COMMON TIME SIGNATURE
common_time_signature = df['time_signature'].mode()[0]
print(f"Most Common Time Signature: {common_time_signature}")
```

```python
explicit_popularity = df.groupby('explicit')['popularity'].mean()
print("Average Popularity for Explicit Tracks:\n", explicit_popularity)
```

```python
# MOST POPULAR ARTIST
artist_popularity = df.groupby('artists')['popularity'].mean().sort_values(ascending=False)
print("\nArtist Popularity (Average):")
print(artist_popularity.head())
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df.index, df['danceability'], label='Danceability', color='green')
plt.plot(df.index, df['energy'], label='Energy', color='red')

plt.title('Line Plot of  Danceability, and Energy')
plt.xlabel('Track Index')
plt.ylabel('Values')

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10, 6))

df_limited = df.iloc[:100]

plt.plot(df_limited.index, df_limited['danceability'], label='Danceability', color='green')
plt.plot(df_limited.index, df_limited['energy'], label='Energy', color='red')

plt.title('Line Plot of  Danceability, and Energy')
plt.xlabel('Track Index')
plt.ylabel('Values')

plt.legend()

plt.show()
```

```python
dance_energy_corr = df[['danceability', 'energy']].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(dance_energy_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Heatmap of Correlation Between Danceability and Energy')
plt.show()
```

```python
correlation_matrix = df[['popularity', 'duration_ms', 'danceability', 'energy',
                         'loudness', 'speechiness', 'acousticness', 'instrumentalness',
                         'liveness', 'valence', 'tempo']].corr()

plt.figure(figsize=(12, 8))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap of Correlation Between Features')
plt.show()
```
